==

        不予许不同类型相加

        不使用显示数据声明类型，且确定一个变量的类型是第一次给他赋值的时候

        一般也是解释性语言，运行代码只需要一个解释器，不需要编译  

    pass 并不会像return一样阻断程序运行，只是一种留白的占位方式。

python 的内存管理
==  
    1. Python的内存管理是由私有heap空间管理的。 
       所有的Python对象和数据结构都在一个私有heap中。程序员没有访问该heap的权限，
       只有解释器才能对它进行操作。为Python的heap空间分配内存是由Python的内存管理模块进行的， 
       其核心API会提供一些访问该模块的方法供程序员使用。Python有自带的垃圾回收系统， 
       它回收并释放没有被使用的内存，让它们能够被其他程序使用。 

       引用计数机制    
       垃圾回收机制 (分代回收）
       内存池机制

    2. getrefcount 本身会引用对象，所以结果大于等于1
    3. python的内存管理机制就是引用计数器机制和垃圾回收机制的混合机制  
       sys.getrefcount()  objgraph   gc.collect()    gc.get_threshold()   分代回收  引用环  
    4. Python有一个私有堆空间来保存所有的对象和数据结构，  
       开发者无法直接访问，它由解释器中的内存管理器管理。  
    5. 循环引用其它对象或引用自全局命名空间的对象的模块，  
       在Python退出时并非完全释放,另外，也不会释放C库保留的内存部分。

Memory Leak：
--
    是指程序中己动态分配的堆内存由于某种原因导致程序未释放或无法释放，
    造成系统内存的浪费，导致程序运行速度减慢甚至系统崩溃等严重后果。
Out of memory
--
    是指程序在申请内存时,没有足够的内存空间供其使用,出现out of memory,
    比如申请了一个integer,但给它存了long才能存下的数,那就是内存溢出

工厂模式
==

```python
class Fruit(object):
    pass

class Apple(Fruit):
    pass

class Banana(Fruit):
    def print_color(self):
        print("banana is in yellow")

class FruitFactory(object):
    fruits = {'apple': Apple, 'banana': Banana}

    def __new__(cls, name):
        if name in cls.fruits.keys():
            cls.fruits[name].__new__(cls.fruits[name])
            return cls.fruits[name]()
        else:
            return Fruit()
```
