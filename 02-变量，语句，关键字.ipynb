### with-上下文管理器
        open 返回文件对象，并调用其__enter_方法，返回值赋给f，然后执行with 语句块，之后执行 文件对象的 __exit__方法。
        open 返回的文件对象实现了__enter__方法 和 __exit__ 方法，而实现了这两个方法的对象称为上下文管理器。
        上下文管理器定义执行 with 语句时要建立的运行时上下文，负责执行 with 语句块上下文中的进入与退出操作
        __enter__ 一般用于资源分配，如打开文件、连接数据库、获取线程锁；在语句体执行前调用，如果有as子句，将子句变量执行其返回值！
        __exit__(exception_type, exception_value, traceback)一般用于资源释放，如关闭文件、关闭数据库连接、释放线程锁。

```python
from contextlib import contextmanager
class WithTest(object):
    @staticmethod
    def test_with():
        # with open(os.path.join(os.path.dirname(__file__), 'test.txt'), 'w') as f:
        #    f.write("李丹阳")
        with open('test.txt', 'w') as f:  # 无论是否出现异常 with 都会关闭文件
            f.write("李雪")

        try:  # 自己关闭异常
            f = open('test.txt', 'w')
            f.write('')
        finally:
            f.close()

    class ContextManager(object):

        # 自定义上下文管理器(环绕通知)
        def __enter__(self):
            print('--------', '__enter__', '--------')

        def __exit__(self, exception_type, exception_value, traceback):
            if exception_type is None:
                print('__exit__', 'without exception')
            else:
                print('__exit__', 'with exception %s:%s:%s' % (exception_type, exception_value, traceback))
                # return False 会重新抛出异常，让外层处理，
                # return True  则不会抛出 （如果没有return 则会当False处理）
                return True

    @staticmethod
    @contextmanager  # 装饰器是有顺序的
    def func():
        try:
            print('__enter__',)
            yield
        finally:
            print('__exit__')
            return True

    @classmethod
    def test_context_manager(cls):
        with cls.ContextManager():
            print('code area1')
            raise (Exception("something error1!"))
            # __exit__ 返回Ture 代码会继续向下执行

        with WithTest.func():
            print('code area2')
            raise (Exception("something error2!"))
```
